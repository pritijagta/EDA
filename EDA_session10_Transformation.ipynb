```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
visa_df=pd.read_csv("Visadataset.csv")
```

```python
exp_data=np.random.exponential(size=1000)
exp_data
```

```python
plt.hist(exp_data,bins=20)
```

```python
normal=np.random.normal(size=10000)
plt.hist(normal)
```

```python
log_data=np.log(exp_data)
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1).hist(log_data,bins=20,label="exponential")
plt.legend()
plt.subplot(1,2,2).hist(exp_data,bins=20,label="logarithmic")
plt.legend()
plt.show()
```

```python
resiprocal_data=np.reciprocal(exp_data)
plt.hist(resiprocal_data,bins=20,label="Reciprocal")
plt.legend()
plt.show()
```

```python
square_root=np.sqrt(exp_data)
plt.hist(square_root)
```

```python

from sklearn.preprocessing import PowerTransformer
power = PowerTransformer(method='yeo-johnson')
data_trans = power.fit_transform(exp_data.reshape(-1,1))
plt.hist(data_trans,bins=10)
plt.show()
```
