### How many Asia people got visa certified and how many asia people got visa denied

```python
continents=list(visa_df['continent'].value_counts().keys())
certified_list=[]
denied_list=[]
for i in continents:
    cond1=visa_df['continent']==i
    cond2=visa_df['case_status']=='Certified'
    cond=cond1 & cond2
    certified_list.append(len(visa_df[cond]))
    print(f"The number of people of {i} got visa certified are {len(visa_df[cond])}")
    
    
    cond1=visa_df['continent']=='Asia'
    cond2=visa_df['case_status']=='Denied'
    cond=cond1 & cond2
    denied_list.append(len(visa_df[cond]))
    print(f"The number of people of {i} got visa Denied are {len(visa_df[cond])}")
pd.DataFrame(zip(certified_list,denied_list),index=continents,columns=['Certified','Denied'])
```
